```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
# Load Bitcoin data into a Pandas DataFrame
df = pd.read_csv('data/BTC-USD.csv', index_col='Date')

# Convert date column to datetime format
df.index = pd.to_datetime(df.index)

# Create a SARIMA model
model = sm.tsa.statespace.SARIMAX(df.Close, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))

# Fit the model to the data
results = model.fit()

```

```python
# Forecast the next 30 days of Bitcoin values
forecast = results.forecast(steps=30)

# Print the forecasted values
print(forecast)
```

```python
plt.plot(forecast)
plt.show()
```
